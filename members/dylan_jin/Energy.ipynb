```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
#Setup
pd.set_option('display.max_rows', 1000)
pd.options.display.float_format = '{:,.2f}'.format
%matplotlib inline
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
energy_all = pd.read_pickle(f'../../datastore/processed_data/not_seasonally_adjusted/1.2 Energy.pickle')
energy_all_std = energy_all.groupby(energy_all.index.year).std()
print(energy_all.head())
print(energy_all.tail())
print(energy_all.info())
print(energy_all.describe())
print(energy_all.series_title.unique())
```

```python
energy_all.value.plot()
plt.title("Energy")
plt.xlabel("Date")
plt.ylabel("Value")
```

```python
energy_all_std.value.plot()
plt.title("Energy Inflation Rate - Yearly STD")
plt.xlabel("Date")
plt.ylabel("Standard Deviation")
```

```python
energy_all['mom_inflation_rate'].plot()
plt.title("Energy MOM Inflation Rate")
plt.xlabel("Date")
plt.ylabel("Inflation Rate")
plt.axhline(y=0, color='black')
```

```python
fig, ax = plt.subplots(nrows=3, figsize=(12, 6))
fig.align_labels()

energy_all['value'].plot(ax=ax[0], x='date', y='value')
energy_all['mom_inflation_rate'].plot(ax=ax[1], x='date', y='mom_inflation_rate')
energy_all_std['value'].plot(ax=ax[2], x='date', y='value')

ax[0].set_title("Energy")
ax[0].set_xlabel("Date")
ax[0].set_ylabel("Value")

ax[1].set_title("MOM Inflation Rate")
ax[1].set_xlabel("Date")
ax[1].set_ylabel("Inflation Rate")
ax[1].axhline(y=0, color='black')

ax[2].set_title("Yearly STD")
ax[2].set_xlabel("Date")
ax[2].set_ylabel("Standard Deviation")

plt.tight_layout(pad=1)
plt.show()
```

```python
energy_all['yoy_inflation_rate'].hist(bins=100)
plt.title("Energy - YOY Inflation Rate")
plt.xlabel("Inflation Rate")
plt.ylabel("Count")
```
